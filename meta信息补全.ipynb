## For 2021, Baolin Liu, Nat Cancer

- Basic meta data from GEO
- Improved information downloaded from supplymentary meterial and in NSCLC_MOESM3_ESM.xlsx

```python
nsclc=pd.read_table("./GSE179994_Tcell.metadata.tsv")
nsclc=nsclc.drop(columns=['celltype'])

nsclc=nsclc.rename(columns={"cluster":"Anno_orig"})
nsclc
```

```python
def write_respond(meta,orig1,orig2,match1,match2,cond1,cond2,cond3):
    respond=list()
    for i in range(nsclc.shape[0]):
        if meta.loc[i][orig1] not in match1:
            respond.append(cond3)
        else:
            if meta.loc[i][orig2] in match2:
                respond.append(cond2)
            else:
                respond.append(cond1)

    return respond
```

```python
match1=["P1","P10","P13","P19","P29",'P30','P33','P35','P36','P37','P38']
match2=['P1.post.3','P13.post.2','P36.post.1','P37.post.1','P38.post.1']
cond1="yes"
cond2="no"
cond3="none"

nsclc_respond=write_respond(nsclc,'patient','sample',match1,match2,cond1,cond2,cond3)
```

```python
nsclc['Respond']= nsclc_respond
```

```python
def write_treatment(meta):
    treat=list()
    for i in range(meta.shape[0]):
        treat.append(meta.iloc[i]['sample'].split(".")[1])

    return treat


nsclc_treat=write_treatment(nsclc)
```

```python
nsclc['treatment']=nsclc_treat
nsclc.columns
```

```python
# 更改列顺序
relist=['cellid',  'sample', 'treatment','Anno_orig', 'Respond' ]
nsclc_new=nsclc[relist]
nsclc_new
```

```python

nsclc_new=nsclc_new.rename(columns={"sample":"patient"})
nsclc_new.to_csv("./nsclc_2021_meta_with_respond.csv",index=False)
```

## Tnbc
- meta_data from supplymetary meterial in tnbc_meta.csv (cut from mmc3.xlsx)
- cell markers in mmc3.xlsx

```python
tnbc=pd.read_csv("./tnbc_meta.csv")
tnbc
```

```python
tnbc=tnbc.rename(columns={"Cluster":"Anno_orig"})
```

```python
def write_treatment(meta):
    treat=list()
    for i in meta.Group:
        treat.append(i.split("-")[0].lower())

    return treat


tnbc_treat=write_treatment(tnbc)
```

```python
def write_respond(meta):
    lis=list()
    for i in meta.Efficacy:
        if i == "PR":
            lis.append('yes')
        else:
            lis.append('no')

    return lis


tnbc_respond=write_respond(tnbc)
```

```python
tnbc=tnbc.drop(columns=['Sample','Origin','Tissue','Number of counts',"Number of genes","Major celltype"])
```

```python
tnbc=tnbc.rename(columns={"Group":"treatment","Treatment":"group","Efficacy":"Respond","Patient":"patient"})
tnbc
```

```python
relist=['cellid',  'patient', 'treatment','Anno_orig', 'Respond','group' ]
tnbc_new=tnbc[relist]
tnbc_new['Respond']=tnbc_respond
tnbc_new['treatment']=tnbc_treat
tnbc_new
```

```python
tnbc_new.to_csv("./tnbc_meta_processed.csv",index=False)
```

## AllCancer Only ImmuneCells 

- 由于因果性的影响，考虑对pre-treatment 但对治疗方法有响应的patient进行分析，对其特异性表达的基因与Respond/nonRespond组里的特异性表达基因进行intersection，可能会更好地确定影响treatment响应的基因

```python
onlyImm=pd.read_csv("./allcancer_only_immuneCell.meta.csv",header=0,index_col=0)
onlyImm
```

```python
patientlist=list(set(onlyImm[(onlyImm.treatment=="post") & (onlyImm.Respond=="yes")].patient))

patientlist
```

```python
def Iftreatment(meta):
    iftreatment=list()
    patient_respond=set(meta[(meta.treatment=="post") & (meta.Respond=="yes")].patient)
    patient_respond.add("P007")

    patient_notreat=set(meta[meta.Respond=="none"].patient)
    
    for index,i in meta.iterrows():
        if (i['patient'] in patient_respond) and (i['treatment']=="pre"):
            iftreatment.append('yes')
        elif i['treatment']=="post" or i['patient'] in patient_notreat:
            iftreatment.append('none')
        else:
            iftreatment.append('no')
    return(iftreatment)

iftreatment=Iftreatment(onlyImm)
```

```python
onlyImm['iftreatment']=iftreatment
onlyImm.to_csv('./onlyImm_withiftreatment.csv',index=False)
```
